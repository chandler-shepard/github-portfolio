```python

age = []
points = []
goals = []
shots = []

for url in sites:

    filein = 'Data/'+url[-13:]
    soup = BeautifulSoup(open(filein), 'lxml')
    
    
    tables = soup.find_all('table')
    #for each in tables:
    #print()
    #print(each)
    
    table = soup.find('table', attrs={'id' : 'skaters'})
    #print(table)

    #find all the rows
    rows = table.find_all('tr', attrs={'class' : ''})
    #print(rows)

    for row in rows:
        playerage = row.find('td', attrs={'class' : 'center', 'data-stat' : 'age'})
        #print(points)
        playerpoints = row.find('td', attrs={'data-stat' : 'points'})
        #print(playerpoints)
        
        
        playergoals = row.find('td', attrs={'data-stat' : 'goals'})
        #print(playergoals)
        
        playershots = row.find('td', attrs={'data-stat' : 'shots'})
        #print(playershots)
    
        if playerage != None and playerage != '':
            newage = playerage.get_text()
            if newage != '':
                age.append(newage)
            

        if playerpoints != None and playerpoints != '':
            newpoints = playerpoints.get_text()
            if int(newpoints) < 200:
                points.append(newpoints)
        
        if playergoals != None and playergoals != '':
            newgoals = playergoals.get_text()
            if int(newgoals) < 150:
                goals.append(newgoals)
        
        if playershots != None and playershots != '':
            newshots = playershots.get_text()
            if int(newshots) < 500 :
                shots.append(newshots)
            
            
            
            
points = [int(i) for i in points]
age = [int(i) for i in age]
goals = [int(i) for i in goals]
shots = [int(i) for i in shots]



#Checking to make sure same number of data points
#print(shots)
#print(goals)
#print(len(shots))
#print(len(goals))



#print(points)
#print(age)
#print(len(points))
#print(len(age))


goals = []
shots = []     

for url in sites:

    filein = 'Data/'+url[-13:]
    soup = BeautifulSoup(open(filein), 'lxml')
    
    
    tables = soup.find_all('table')
    #for each in tables:
    #print()
    #print(each)
    
    table = soup.find('table', attrs={'id' : 'skaters'})
    #print(table)

    #find all the rows
    rows = table.find_all('tr', attrs={'class' : ''})
    #print(rows)

    for row in rows:
        playergoals = row.find('td', attrs={'data-stat' : 'goals'})
        #print(playergoals)
        
        playershots = row.find('td', attrs={'data-stat' : 'shots'})
        #print(playershots)
        
    if playergoals != None and playergoals != '':
        newgoals = playergoals.get_text()
        if int(newgoals) < 150:
            goals.append(newgoals)
        
    if playershots != None and playershots != '':
        newshots = playershots.get_text()
        if int(newshots) < 500 :
            shots.append(newshots)

goals = [int(i) for i in goals]
shots = [int(i) for i in shots]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 

#Plot Age vs Points
plt.figure(figsize=(8,8))
plt.rc('text', usetex=True)
plt.rc('font', size=24, **{'family':'DejaVu Sans','sans-serif':['Helvetica']})
plt.rcParams['xtick.major.pad'] = 8
plt.rcParams['ytick.major.pad'] = 8

title = "Age vs Pts"
plt.title(title, fontsize= 30)

plt.xlabel('Age of Player')
plt.ylabel('Points in NHL Season')




plt.scatter(age,points, label='Normal', s=100, alpha=0.5, linewidth=0)

#Plot Goals and Shots on Goal 
plt.figure(figsize=(8,8))
plt.rc('text', usetex=True)
plt.rc('font', size=24, **{'family':'DejaVu Sans','sans-serif':['Helvetica']})
plt.rcParams['xtick.major.pad'] = 8
plt.rcParams['ytick.major.pad'] = 8

title = "Goals vs Shots on Goal"
plt.title(title, fontsize= 30)

plt.xlabel('Goals')
plt.ylabel('Shots on Goal')




plt.scatter(goals,shots, label='Normal', s=100, alpha=0.5, linewidth=0)
```
