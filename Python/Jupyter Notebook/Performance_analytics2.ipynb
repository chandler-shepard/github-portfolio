This next step is to go through the urls comtaining all the teams stats from the set season. We do this by opening the url reading in the contents, writing them in a local file in the folder Data and then closing the file.

```python
for url in sites:
    #print ('Download from :', url)
    #dowlnload
    filehandle = urllib.request.urlopen(url)
    html = filehandle.read()



    fileout = 'Data/'+url[-13:].replace('/', '_')
    #print ('Save to : ', fileout, '\n')

    #save file to disk
    f = open(fileout,'w')
    f.write(html.decode('utf-8'))
    f.close()
print("done")
```

In this section we open each file with the team stats page and mine through the contents to create variables to hold the age, powerplay goals and penalty minutes for all players on the team. This is done for all teams thats played that season of Hockey. 

```python
age = []
ppgoals = []
PIM = []

for url in sites:
#     print(url)
    filein = 'Data/'+url[-13:].replace('/', '_')
    soup = BeautifulSoup(open(filein), 'lxml')
    
    
    tables = soup.find_all('table')
    #for each in tables:
    #print()
    #print(each)
    
    table = soup.find('table', attrs={'id' : 'skaters'})
    #print(table)

    #find all the rows
    rows = table.find_all('tr', attrs={'class' : ''})
    #print(rows)

    for row in rows:
        playerage = row.find('td', attrs={'class' : 'center', 'data-stat' : 'age'})
        #print(playerage)
        
        playerPIM = row.find('td', attrs={'data-stat' : 'pen_min'})
        #print (playerPIM)
        
        
        playerppgoals = row.find('td', attrs={'class' : 'right', 'data-stat' : 'goals_pp'})
        #print (playerppgoals)
        
      
        control_age = 0
        control_PIM = 0 
        control_ppgoals = 0
        
        
    
        if playerage != None and playerage != '':
            newage = playerage.get_text()
            if newage != '':
                #age.append(newage)
                control_age = 1 
            

        if playerPIM != None and playerPIM != '':
            newPIM = playerPIM.get_text()
            if int(newPIM) < 100:
                #PIM.append(newPIM)
                control_PIM = 1
        
        if playerppgoals != None and playerppgoals != '':
            newplayerppgoals = playerppgoals.get_text()
             #print(newplayerppgoals)
            if int(newplayerppgoals) < 50:
                #ppgoals.append(newplayerppgoals)
                control_ppgoals = 1
        
      
        if control_age == 1 and control_PIM == 1 and control_ppgoals == 1:
            age.append(int(newage))
            PIM.append(int(newPIM))
            ppgoals.append(int(newplayerppgoals))
```

```python
def measure_probability_distribution (winners_points):
    
    average_value = 0.0
    variance = 0.0
    
    pdf = {}
    norm = 0.0
    
    ##count number of observations
    for x in winners_points:
        if x not in pdf:
            pdf[x] = 0.0
        pdf[x] += 1.0
        norm += 1.0
        
        average_value += x
        variance += x*x
        
        
    average_value /= norm
    variance /= norm
    variance = variance - average_value * average_value
        
        
    ##normalize pdf
    for x in pdf:
        pdf[x] /= norm
    
    
    return pdf, average_value, variance
#####################################

#measure_probability_distribution(age)
#measure_probability_distribution(ppgoals)
```

```python
#Histogram Age of players

fig = plt.figure()
fig.set_size_inches(10,5)
plt.rc('text', usetex=True)
plt.rc('font', size=22, **{'family':'DejaVu Sans','sans-serif':['Helvetica']})
plt.rcParams['xtick.major.pad'] = 8
plt.rcParams['ytick.major.pad'] = 8


plt.hist(age, bins = 15, edgecolor="black")
plt.title ("Age of NHL Players in " + season)
plt.xlabel("Age")
plt.ylabel("Number of Players")

plt.show()
```

```python
#Scatterplot Penalty in Minuts vs Power Play Goals

fig = plt.figure()
fig.set_size_inches(10,5)
plt.rc('text', usetex=True)
plt.rc('font', size=22, **{'family':'DejaVu Sans','sans-serif':['Helvetica']})
plt.rcParams['xtick.major.pad'] = 8
plt.rcParams['ytick.major.pad'] = 8

title = "Penalty in Minutes vs Power Play Goals"
plt.title(title, fontsize= 30)

plt.xlabel('Penalty in Minutes')
plt.ylabel('Power Play Goals')



plt.scatter(PIM,ppgoals, label='Normal', s=100, alpha=0.5, linewidth=0)
plt.show()
```
